# B. Introduction to Visualization with Matplotlib

Data visualization is the graphical representation of data or information. It plays a pivotal role in data analytics as it helps to express and deliver data in an easy-to-understand manner.

Matplotlib is the most basic but important Python library for data visualization, and in this lesson, we will be learning how to draw plain graphs with it.

### _Objective_
- **Matplotlib**: Learning how to draw plain graphs on your data using `matplotlib.pyplot`. 

# \[1. Drawing Linear Graphs\]

## 1. Preparing Matplotlib

When using `matplotlib` to draw plots in the Jupyter notebook, you need to declare `%matplotlib inline` in the beginning  
so that graphs can be displayed in the frontend.

In this section, we'll be using `matplitlib.pyplot` among matplotlib libraries to draw basic graphs on our data.<br>
Let's import `matplotlib.pyplot` under an alias `plt`

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 2. Drawing Basic Graphs
The process of drawing a plot with `plt.plot` can be divided into 3 stages as follows.
- `plt.figure()` to create a new figure object before drawing.
- `plt.plot()` to draw a plot in the figure object.
- `plt.show()` prints the plot

#### Example Data) The transition of student's subjects score
The example data are the academic records of a student over the past 3 years.

```python
scores = np.array([
    [83, 85, 84, 81, 80, 85, 84, 81, 79, 75, 76, 74],
    [86, 90, 91, 89, 92, 90, 87, 85, 87, 84, 82, 80],
    [73, 78, 82, 87, 92, 96, 95, 93, 94, 96, 97, 98]])
scores_df = pd.DataFrame(scores, 
                         columns = ['1-1mid', '1-1final', '1-2mid', '1-2final', 
                                    '2-1mid', '2-1final', '2-2mid', '2-2final',
                                    '3-1mid', '3-1final', '3-2mid', '3-2fianl'],
                        index = ['Science', 'English', 'Math'])
display(scores_df)
```

Assiging all science scores to an array named `science_score`<br><br><br><br>

```python
science_score = scores[0] # science_score holds all students' science scores
science_score
```

When plotting `science_score`, you will get a result as shown below.

```python
# (1) Create a figure object for data visualization. The size of the figure is (12, 5)
fig = plt.figure(figsize=(12, 5))

# (2) Draw a graph of students' science scores over 3 years
plt.plot(science_score) 
# x-axis: data index (Index)
# y-axis: data value (Value)

# (3) Print the plot
plt.show()
```

Despite the graph looks okay, there seems to be something missing. 

First, there is no information on what the graph is about and what each axis indicates.  
To tackle these issues, pyplot provides a number of methods you can use for customizing the graph elements. 

Let's put some more information to the graph using them. 

## 2. Adding titles to the figure

With `plt.title()`, you can set the title to the figure.

```python
science_score = scores[0]
fig = plt.figure(figsize=(12, 5))
plt.plot(science_score) 
plt.title("Changes in science scores") # setting title to the figure 
plt.show()
```

## 3. Setting values for the x-axis.

The x-axis values consist of numbers from 0 to 11 as many as the number of columns.  
However, each value should rather correspond to the times of events, in this case, when the exams took place. 

So, let's change the x-axis as follows.

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"] # assigning the list of x-axis values to `xs`

science_score = scores[0] 
fig = plt.figure(figsize=(12, 5)) 

plt.plot(xs, science_score) # passing `xs` to `x` parameter when drawing a plot to set it as x-axis labels
plt.title("Changes in Science Scores")
plt.show()
```

## 4. Setting labels for axes
Now, we'll set labels to each axis, the x-xis with "Semester" and y-axis with "Score"
- `plt.xlabel()` to set the x-axis label
- `plt.ylabel()` to set the y-axis label

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

science_score = scores[0]

fig = plt.figure(figsize=(12, 5)) 

plt.plot(xs, science_score)

plt.title("Changes in Science Scores")
plt.xlabel("Semester") # labeling the x-axis as "Semester"
plt.ylabel("Score") # labeling the y-axis as "Score"
plt.show()
```

## 5. Adding a New Plot to the Existing Plots

Suppose that we just came up with a new idea to compare changes in students' science score over the past 3 years with those in math,  
so we will draw an additional plot for the math scores.

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

science_score = scores[0] # getting students' science scores in an array
math_score = scores[2] # getting students' math scores in an array

fig = plt.figure(figsize=(12, 5)) # creating a figure object of size (12, 5)

plt.plot(xs, science_score) # drawing the first plot for changes in science scores
plt.plot(xs, math_score) # adding the second plot to the existing plot object for comparison
plt.title("Changes in math/science scores")
plt.xlabel("Semester")
plt.ylabel("Score")
plt.show()
```

## 6. Adding a legend to the figure

If you have two or more plots on a single figure object, it would be difficult to know which plot is what.

In order to provide information on what readers are looking at, we'll add a legend to the figure using `plt.legend()`.  
The legend tells us what each plot represents.

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

science_score = scores[0]
math_score = scores[2]

fig = plt.figure(figsize=(12, 5))

plt.plot(xs, science_score)
plt.plot(xs, math_score)

plt.title("Changes in Math/Science Scores")
plt.xlabel("Semester")
plt.ylabel("Score")

plt.legend(["Science", "Math"]) # Since we have 2 plots, it's essential to provide information on what is what.
plt.show()
```

## 7. customizing ticks
If you want to display only specific ticks on each axis of the graph, you can customize the tick values with `plt.xticks` and `plt.yticks`.  
**This would be useful especially when ticks or axis labels are too tightly placed to read,  
or in case you have specific ticks you want to emphasize on**.

For the example data, you can display the starting and ending points of the plots as well as the point of time there was a radical change in your scores.

-`plt.xticks(ticks, [labels])`- used for customizing x-axis ticks and labels  
-`plt.yticks(ticks, [labels])`- used for customizing y-axis ticks and labels 

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

science_score = scores[0]
math_score = scores[2]

fig = plt.figure(figsize=(12, 5))

plt.plot(xs, science_score)
plt.plot(xs, math_score)

plt.title("Changes in Math/Science scores")
plt.xlabel("Semester")
plt.ylabel("Score")

plt.legend(["Science", "Math"])

plt.xticks(["2-1final"]) # `2-1final`to mark the period when the scores rose at the most
plt.yticks([85])         # `85` 
plt.show()
```

## 8. Setting value limits 

When there are two or more plots on a single figure, it will difficult to compare one with one another as their value ranges vary.<br>
In this case, you can use `plt.xlim()` and `plt.ylim()` to set the value ranges for each axis

- `plt.xlim(left, right)` -  sets the minimum(left) and maximum value(right) that can be plotted on the x-axis of the graph
- `plt.ylim(left, right)` -  sets the minimum(left) and maximum value(right) that can be plotted on the y-axis of the graph

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

science_score = scores[0]
math_score = scores[2]

fig = plt.figure(figsize=(12, 5))

plt.plot(xs, science_score)
plt.plot(xs, math_score)

plt.title("Science/Math Transition")
plt.xlabel("Semester")
plt.ylabel("Score")

plt.legend(["Science", "Math"])

plt.xticks(["2-1final"])

plt.ylim(70, 100) # limiting y-axis values ranges between 70 and 100

plt.show()
```

## 9. Drawing multiple plots at once

If there are too many subjects to be plotted, it will be burdensome to draw them one by one.  
Instead, `matplotlib.pyplot` provides tools you can draw multiple lines with at a single execution.

```python
scores 
```

### (1) Reshaping the data

When visualizing the above array, you should first reshape the data.

We're going to transpose the data with `np.transpose`.

- `np.transpose()`: flips a matrix over its diagonal

```python
transposed_scores = np.transpose(scores)
transposed_scores 
```

### (2) Drawing plots
Once the array has been reshaped, you can pass the data to `plt.plot()` in the same way as before.

```python
xs = ["1-1mid","1-1final","1-2mid","1-2final",
      "2-1mid","2-1final","2-2mid","2-2final",
      "3-1mid","3-1final","3-2mid","3-2final"]

fig = plt.figure(figsize=(12, 5)) 

plt.plot(xs, transposed_scores) 
  
plt.title("Subjects Transition") # title of the figure
plt.xlabel("Semester")
plt.ylabel("Score")

plt.legend(["Science","English","Math"]) 
plt.xticks(["2-1final"]) # only a single tick will be displayed on x-aixs

plt.ylim(70,100)
plt.show()
```
